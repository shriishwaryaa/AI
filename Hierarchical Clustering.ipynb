```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```
data = pd.read_csv('/content/Wholesale customers data.csv')
data.head()
```

```
from sklearn.preprocessing import normalize
dataNormalized = normalize(data)
dataNormalized = pd.DataFrame(dataNormalized, columns=data.columns)
dataNormalized.head()
```

```
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10, 7))
plt.title('Dendrograms')
dend = shc.dendrogram(shc.linkage(dataNormalized, method='ward'))
```

```
plt.figure(figsize=(10, 7))
plt.title('Dendrograms')
dend = shc.dendrogram(shc.linkage(dataNormalized, method='ward'))
plt.axhline(y=5, color='y', linestyle='--')
```

```
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
classes = cluster.fit_predict(dataNormalized)
print(classes)
```

```
plt.figure(figsize=(10, 7))  
plt.scatter(dataNormalized['Milk'], dataNormalized['Grocery'], c=classes) 
```

```
plt.figure(figsize=(10, 7))  
plt.scatter(dataNormalized['Frozen'], dataNormalized['Detergents_Paper'], c=cluster.labels_) 
```
